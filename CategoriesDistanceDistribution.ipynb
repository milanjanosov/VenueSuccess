```python
import os 
import numpy as np
import mpu
import pandas as pd
import math
from collections import Counter
```

```python
city    = 'london'
infile  = '../ProcessedData/london/venues_info/venues_all_categories_times.dat'
infile2 = '../ProcessedData/london/venues_info/venues_ward_full.dat'
```

```python
venues_categories = {}

for ind, line in enumerate(open(infile)):
    
    #if ind == 10: break      
    fields = line.strip().split('\t')
    venue  = fields[1]
    cat    = fields[2]
    
    if venue not in venues_categories:
        venues_categories[venue] = cat
    
```

```python
all_categories = set(venues_categories.values())
all_cats_count = Counter(venues_categories.values())
all_venues     = len(venues_categories)
all_categories
```

```python
veunes_wards  = {}
venues_coords = {}
wards_venues  = {}

for ind, line in enumerate(open(infile2)):

    if 'venue' not in line:
    
        #ifind == 10: break
        fields = line.strip().split('\t')
        venue  = fields[0]
        lng    = float(fields[1])
        lat    = float(fields[2])
        ward   = fields[3]
        
        veunes_wards[venue]  = ward
        venues_coords[venue] = (lng, lat)
        
        if ward not in wards_venues:
            wards_venues[ward] = [venue]
        else:
            wards_venues[ward].append(venue)
        
```
