# HELPERS AND PARAMS

```python
def get_centroids(c):
     return (sum(c[0])/len(c[0]),sum(c[1])/len(c[1]))

def encode_cats(df, column):
    le = preprocessing.LabelEncoder()
    le.fit(df[column])
    df[column] = le.transform(df[column])    
    return df
```

### Ward Structural Stats 

```python
ward_stats_f = datafile + '/venues_ward_full.dat'
ward_stats   = pd.read_csv(ward_stats_f, sep = '\t', index_col = 0)
ward_stats   = ward_stats[~ward_stats.index.duplicated(keep='first')]   
ward_stats   = ward_stats.rename(columns = {'ward' : 'GSS_CODE'})#.drop(['ward'], axis=1)

# get the distance from Trafalgar square
london_lng, london_lat = (-0.128069, 51.508039)
lngs = list(ward_stats.lng)
lats = list(ward_stats.lat)


ward_region = pd.read_csv('FINAL_DATA/london_wards_borough.csv', sep = '\t', index_col = 0)

ward_stats['venue'] = list(ward_stats.index)
ward_stats = ward_stats.merge(ward_region, on = 'GSS_CODE', how='right')

ward_stats['distance_from_center'] = ward_stats.apply(lambda row: mpu.haversine_distance((row.lat, row.lng ), (london_lat, london_lng)), axis=1)

ward_stats.index = ward_stats.venue
ward_stats = ward_stats.drop(columns = ['venue', 'length', 'lng', 'lat', 'lng1', 'lat1', 'lng0', 'lat0','area'])
ward_stats = ward_stats.rename(columns = { 'distance_from_center' : 'WHERE_distance_from_center'})
ward_stats = ward_stats.drop(columns = ['GSS_CODE', 'BOROUGH'])

print ward_stats.loc['4c4b196342b4d13a9344337e'].WHERE_sub_region
print ward_stats.loc['533479f0498e10485ff2d132'].WHERE_sub_region
print ward_stats.loc['542aeb3b498e5227edba2428'].WHERE_sub_region
print ward_stats.loc['54982bcb498eea3f58ba08f7'].WHERE_sub_region
print ward_stats.loc['54ca4b13498e65e160dcb930'].WHERE_sub_region
print ward_stats.loc['550d9b05498e2c34ab179188'].WHERE_sub_region
print ward_stats.loc['5545f4c9498e1b710b5f2dce'].WHERE_sub_region



encode_cats(ward_stats, 'WHERE_sub_region')







 
```

### Demography

```python
wards = pd.read_csv(datafile + '/venues_ward_full.dat', sep = '\t', index_col = 0)
wards = wards[~wards.index.duplicated(keep='first')]   
wards = wards.drop(columns = [w for w in wards.keys() if 'ward' not in w])
wards['venue'] = wards.index

wards.head()
```

```python
demog_f = datafile + '/london_wards_demographics.csv'
demog   = pd.read_csv(demog_f, sep = ',', index_col = 0)
demog   = demog.rename(columns = { k : 'WHERE_' + k for k in demog.keys()})
demog   = demog.rename(columns = {'WHERE_GSS_CODE' : 'ward'})
demog   = wards.merge(demog, on = 'ward', how = 'right')

demog.index = demog.venue
demog = demog.drop(columns = ['venue', 'ward'])
demog.head()

print demog.keys()
```

### Categories features

```python
# category stats of wards: number of venues, fraction of food, nightlife..., entropy
ward_cats_f  = datafile + '/london_WARD_category_stats.csv'
ward_cats    = pd.read_csv(ward_cats_f, sep = '\t', index_col = 0)
ward_cats    = ward_cats[~ward_cats.index.duplicated(keep='first')]   
ward_cats    = encode_cats(ward_cats, 'venue_cat')
ward_cats    = encode_cats(ward_cats, 'venue_subcat')

ward_cats.index.names = ['venue']

ward_cats = ward_cats.rename(columns = {k : 'WHERE_' + k for k in ward_cats.keys()})
ward_cats = ward_cats.rename(columns = {'WHERE_venue_cat' : 'WHAT_venue_cat', 'WHERE_venue_subcat' : 'WHAT_venue_subcat'})


vens_money = pd.read_csv('FINAL_DATA/london_venues_4sq_money.dat', sep = ',', index_col = 0)
vens_money = vens_money.replace('na', '0')
ward_cats['WHAT_pricerange'] = pd.to_numeric(vens_money.money)

ward_cats.head()
```

### Geographical spans

```python
# user and venue span measures
venue_span_f       = datafile + '/london_venues_geo_features.csv'
venues_user_span_f = datafile + '/london_users_AGGERG_venues_geo_features.csv'
venue_span         = pd.read_csv(venue_span_f, sep = '\t', index_col = 0).replace([np.inf, -np.inf], np.nan).fillna(0.0)
venues_user_span   = pd.read_csv(venues_user_span_f, sep = '\t', index_col = 0).replace([np.inf, -np.inf], np.nan).fillna(0.0)
venue_span         = venue_span.rename( columns = { k : 'v_' + k for k in venue_span.keys()})
venues_user_span   = venues_user_span.rename( columns = { k : 'u_' + k for k in venues_user_span.keys()})

geo_spans = venue_span.join(venues_user_span)
geo_spans.index.name = 'venue'
geo_spans   = geo_spans.rename(columns = { k : 'WHO_' + k for k in geo_spans.keys()})
geo_spans   = geo_spans.drop(columns = [k for k in geo_spans.keys() if 'polygon' in k])

geo_spans.head()
```
