```python
city      = 'london'
outfolder = '../ProcessedData/' + city + '/'
```

```python
outname = 'users_similarity'
outfile_DF = outfolder +  'networks/gephi/' + 'COMPARE_DF_thresholds_london_'+outname+'.dat' #COMPARE_NC_thresholds_' + outname + '.dat'
outfile_NC = outfolder +  'networks/gephi/' + 'COMPARE_NC_thresholds_london_'+outname+'.dat' #COMPARE_NC_thresholds_' + outname + '.dat'
            
```

```python
ps_DF, edges_DF, nodes_DF = zip( * [ [float(fff) for fff in line.strip().split('\t')] for line in open(outfile_DF) ] )
ps_NC, edges_NC, nodes_NC = zip( * [ [float(fff) for fff in line.strip().split('\t')] for line in open(outfile_NC) ] )

```

```python
import matplotlib.pyplot as plt
```

```python
f, ax  = plt.subplots(2, 2, figsize=(12, 8))


ax[0,0].plot(ps_NC, edges_NC, 'bo', label = 'Noise-Corrected (NC)')
ax[0,0].set_yscale('log')
ax[0,1].plot(ps_NC, nodes_NC, 'bo', label = 'Noise-Corrected (NC)')

ax[0,0].set_ylabel('Number of edges', fontsize = 16)
ax[0,1].set_ylabel('Number of nodes', fontsize = 16)
ax[0,0].set_xlabel('Weight param', fontsize = 16)
ax[0,1].set_xlabel('Weight param', fontsize = 16)
ax[0,0].legend(loc = 'left', fontsize = 15)
ax[0,1].legend(loc = 'left', fontsize = 15)



ax[1,0].plot(ps_DF, edges_DF, 'bo', label = 'Disparity filter (DF)')
ax[1,0].set_yscale('log')
ax[1,1].plot(ps_DF, nodes_DF, 'bo', label = 'Disparity filter (DF)')

ax[1,0].set_ylabel('Number of edges', fontsize = 16)
ax[1,1].set_ylabel('Number of nodes', fontsize = 16)
ax[1,0].set_xlabel('Significance param', fontsize = 16)
ax[1,1].set_xlabel('Significance param', fontsize = 16)
ax[1,0].legend(loc = 'left', fontsize = 15)
ax[1,1].legend(loc = 'left', fontsize = 15)
```
